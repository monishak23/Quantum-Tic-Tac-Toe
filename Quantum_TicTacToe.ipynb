```python
from qiskit import *
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_bloch_multivector
```

```python
def table():
    print('********** This the table state ******************')
    inx =0 
    while(inx < (len(l_initial_ordered))):
          print('-----\n'+str(l_initial_ordered[inx])+'|'+str(l_initial_ordered[inx+1])+'|'+str(l_initial_ordered[inx+2])+'|'+str(l_initial_ordered[inx+3]))
          inx = inx+4
```

```python
l_initial_ordered = ['*','*','*','*','*','*','*','*','*','*','*','*','*','*','*','*']
table()
```

```python
qc = QuantumCircuit(16, 16)
qc.name = "Tic toc toe"
```

```python
max_choice = 16
n_player = 0 
count = 0
qubit = 0 
control =0 
target =0

while(count < max_choice ):

    ch = input('Choose between move and entanglement of bits : ')

    if(n_player%2 ==0 and ch=='move'):
          qubit = input('Choose a qubit ')
          qubit = int(qubit)
          qc.measure(qubit,qubit)
          l_initial_ordered[qubit] = '0'
          count = count+1
          table()
          
    if(ch=='entangle'):
          if(count >= max_choice-1) :
            table()
            print("No possible qubits to entangle..........Choose move")
            
          else:
              control = input('Enter control qubit for entanglement :')
              control1 = int(control)
              target = input('Enter target qubit for entanglement :')
              target1 = int(target)
              qc.x(target1)
              qc.h(control1)
              qc.cx(control1,target1)
              qc.measure(control1,control1)
              qc.measure(target1,target1)
              l_initial_ordered[control1] = 'E'
              l_initial_ordered[target1] = 'E'
              count = count+2
              table()
                
                
    if(n_player%2 !=0 and ch=='move'):
          qubit = input('Choose a qubit ')
          qubit = int(qubit)
          qc.x(qubit)
          qc.measure(qubit,qubit)
          l_initial_ordered[qubit] = '1'
          count = count+1
          table()
            
            
   
    n_player = n_player+1

```

```python
qc.draw('text')
```

```python
job = execute(qc, BasicAer.get_backend('qasm_simulator'), shots=1000)
result = job.result()
l_final_ordered=list(map(lambda x: int(x),list(list(result.get_counts().keys())[0][::-1])))
print('The final state is:')
inx =0 
while(inx < (len(l_initial_ordered))):
          print('-----\n'+str(l_final_ordered[inx])+'|'+str(l_final_ordered[inx+1])+'|'+str(l_final_ordered[inx+2])+'|'+str(l_final_ordered[inx+3]))
          inx = inx+4
print('-----')
```

```python
def check_winner(board,mark):

    return(((board[0]==mark) and (board[1]== mark) and (board[2]==mark) and (board[3]==mark) )or #for row1 

            ((board[4]==mark) and (board[5]==mark) and (board[6]==mark) and (board[7]==mark)  )or #for row2

            ((board[8]==mark) and (board[9]==mark) and (board[10]==mark) and (board[11]==mark) )or #for row3
           
            ((board[12]==mark) and (board[13]==mark) and (board[14]==mark) and (board[15]==mark) )or #for row4

            ((board[0]==mark) and (board[4]==mark) and (board[8]== mark) and (board[12]== mark) )or#for Colm1 

            ((board[1]==mark) and (board[5]==mark) and (board[9]==mark) and (board[13]== mark) )or #for Colm 2

            ((board[2]==mark) and (board[6]==mark) and (board[10]==mark) and (board[14]== mark))or #for colm 3
           
            ((board[3]==mark) and (board[7]==mark) and (board[11]==mark) and (board[15]== mark))or #for colm 4

            ((board[0]==mark) and (board[5]==mark) and (board[10]==mark) and (board[15]==mark) )or #diagonal 1

            ((board[3]==mark) and (board[7]==mark) and (board[9]==mark) and (board[12]==mark))) #diagonal 2
```

```python

if (check_winner(l_final_ordered,0) ):## to check if player 1 won
    print('Player 1 won!')

elif (check_winner(l_final_ordered,1)): ## to check if player 2 won
    print('Player 2 won!')
    
else:
    print("Play the game again!!!!!!!!!")

```
